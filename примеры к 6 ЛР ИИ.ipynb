```python
import pandas as pd
import numpy as np
data = {
"Имя": ["Анна", "Иван", "Ольга", "Петр", "Мария", "Дмитрий",
"Елена"],
"Баланс на счете": [50000, 60000, 45000, 70000, 65000, 400000,
450000]
}
df = pd.DataFrame(data)
print(df)
```

```python
Q1 = df["Баланс на счете"].quantile(0.25)
Q3 = df["Баланс на счете"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df_removed = df[(df["Баланс на счете"] >= lower) & (df["Баланс на счете"] <= upper)]
print(df_removed)
```

```python
df_clipped = df.copy()
df_clipped["Баланс на счете"] = df["Баланс на счете"].clip(lower,
upper)
print(df_clipped)
```

```python
median = df["Баланс на счете"].median()
df_replaced = df.copy()
df_replaced["Баланс на счете"] = df["Баланс на счете"].apply(
lambda x: median if x < lower or x > upper else x
)
print(df_replaced)
```

```python
df_log = df.copy()
df_log["Лог баланс"] = np.log1p(df["Баланс на счете"])
print(df_log[["Имя", "Баланс на счете", "Лог баланс"]])
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]
for col in ["carat", "depth", "table", "x", "y", "z"]:
    df = remove_outliers_iqr(df, col)
df.head()
```
